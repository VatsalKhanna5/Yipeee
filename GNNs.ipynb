# Graph Neural Networks (GNNs) in Practice

In many real problems, data is not just a list of independent samples.
Instead, things are *connected*:

- Papers citing each other (citation networks)
- Users connected via friendships (social networks)
- Molecules with atoms and bonds (chemical graphs)
- Web pages with hyperlinks (web graph)

A **Graph Neural Network (GNN)** is a neural network that works **directly on graphs**.
The core idea is:

> Each node updates its representation by **aggregating information from its neighbors**.

This “message passing” happens layer by layer, so information flows through the graph.
